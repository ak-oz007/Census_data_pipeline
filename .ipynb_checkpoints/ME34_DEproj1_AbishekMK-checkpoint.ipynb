# Task 2: Rename the State names

```python
#Using list comprehension to capitalize words except 'and'
def state_rename(sname):
    words = sname.split()
    renamed_state = [word.capitalize() if word.lower() != 'and' else word.lower() for word in words]
    return ' '.join(renamed_state)
```

```python
df['State/UT'] = df['State/UT'].apply(state_rename)
```

```python
df.head(10)
```

# Task 3: New State/UT Formation

```python
#Allocating Kargil and Leh to Ladakh
def newState_formation(row):
    if row['State/UT'] == 'Jammu and Kashmir' and row['District'] in ['Leh(Ladakh)','Kargil']:
        return 'Ladakh'
    return row['State/UT']
```

```python
df['State/UT'] = df.apply(newState_formation, axis=1)
```

# Query 13: Average household Size(State)

```python
#Total people / Total households = Avg size
#Total number of people is calculated based on Assumptions for 7 and 9
query13 = """
SELECT `State/UT`,
       (SUM(Household_size_1_person_Households) * 1 +
        SUM(Household_size_2_persons_Households) * 2 +
        SUM(Household_size_3_persons_Households) * 3 +
        SUM(Household_size_4_persons_Households) * 4 +
        SUM(Household_size_5_persons_Households) * 5 +
        SUM(Household_size_6_8_persons_Households) * 7 +  
        SUM(Household_size_9_persons_and_above_Households) * 9)  
        /
       (SUM(Household_size_1_person_Households) +
        SUM(Household_size_2_persons_Households) +
        SUM(Household_size_3_persons_Households) +
        SUM(Household_size_4_persons_Households) +
        SUM(Household_size_5_persons_Households) +
        SUM(Household_size_6_8_persons_Households) +
        SUM(Household_size_9_persons_and_above_Households)) as Average_Household_Size
FROM census
GROUP BY `State/UT`
"""
cursor.execute(query13)
results = cursor.fetchall()
```

```python
for row in results:
    print(f"State/UT : {row[0]} |Average Household Size : {row[1]: .0f}")
```

```python
query18 = """
SELECT `State/UT`,
       (SUM(Married_couples_1_Households) / SUM(Married_couples_1_Households + Married_couples_2_Households + Married_couples_3_Households + Married_couples_3_or_more_Households + Married_couples_4_Households + Married_couples_5__Households + Married_couples_None_Households)) * 100 as One_Households,
       (SUM(Married_couples_2_Households) / SUM(Married_couples_1_Households + Married_couples_2_Households + Married_couples_3_Households + Married_couples_3_or_more_Households + Married_couples_4_Households + Married_couples_5__Households + Married_couples_None_Households)) * 100 as Two_Households,
       (SUM(Married_couples_3_Households) / SUM(Married_couples_1_Households + Married_couples_2_Households + Married_couples_3_Households + Married_couples_3_or_more_Households + Married_couples_4_Households + Married_couples_5__Households + Married_couples_None_Households)) * 100 as Three_Households,
       (SUM(Married_couples_3_or_more_Households) / SUM(Married_couples_1_Households + Married_couples_2_Households + Married_couples_3_Households + Married_couples_3_or_more_Households + Married_couples_4_Households + Married_couples_5__Households + Married_couples_None_Households)) * 100 as Three_or_More_Households,
       (SUM(Married_couples_4_Households) / SUM(Married_couples_1_Households + Married_couples_2_Households + Married_couples_3_Households + Married_couples_3_or_more_Households + Married_couples_4_Households + Married_couples_5__Households + Married_couples_None_Households)) * 100 as Four_Households,
       (SUM(Married_couples_5__Households) / SUM(Married_couples_1_Households + Married_couples_2_Households + Married_couples_3_Households + Married_couples_3_or_more_Households + Married_couples_4_Households + Married_couples_5__Households + Married_couples_None_Households)) * 100 as Five_Households,
       (SUM(Married_couples_None_Households) / SUM(Married_couples_1_Households + Married_couples_2_Households + Married_couples_3_Households + Married_couples_3_or_more_Households + Married_couples_4_Households + Married_couples_5__Households + Married_couples_None_Households)) * 100 as None_Households
FROM census
GROUP BY `State/UT`
"""
cursor.execute(query18)
results = cursor.fetchall()
```

```python
for row in results:
    print(f"State/UT: {row[0]} |1 Household: ({row[1]:.2f}%) |2 Households: ({row[2]:.2f}%) |3 Households: ({row[3]:.2f}%) |3 or More Households: ({row[4]:.2f}%) |4 Households: ({row[5]:.2f}%) |5 Households: ({row[6]:.2f}%) |None Households: ({row[7]:.2f}%)")
```

# Query 20: Literacy percentage(State)

```python
query20 = """
SELECT `State/UT`,
(SUM(Literate) / SUM(Population)) * 100 as Literacy_precent
FROM census
GROUP BY `State/UT`
"""
cursor.execute(query20)
results = cursor.fetchall()
```
